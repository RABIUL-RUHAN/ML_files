```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset =[45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]
```

```python
range_of_data = max(dataset) - min(dataset)
range_of_data
```

```python
print("Variance:", np.var(dataset))
```

```python
print("Standard Deviation:", np.std(dataset))
```

```python
#2nd quartile (Q2) / Median
print("Median (Q2):", np.percentile(dataset, 50))
```

```python
# 1st and 3rd quartile (Q3)
Q1 = np.percentile(dataset, 25)
Q2 = np.percentile(dataset, 75)
print("1st Quartile (Q1):", Q1)
print("3rd Quartile (Q3):", Q2)
IQR = Q2 - Q1
print("Interquartile Range (IQR):", IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q2 + 1.5 * IQR
print("Lower Fence:", lower_bound)
print("Upper Fence:", upper_bound)
```

```python
cnt=0
for value in dataset:
    if value < lower_bound or value > upper_bound:
        print(f"Outlier detected: {value}")
        cnt += 1
print(f"Total outliers detected: {cnt}")
```
